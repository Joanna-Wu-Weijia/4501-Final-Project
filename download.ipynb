```python
import re
import requests
from bs4 import BeautifulSoup
import os
```

```python
url = 'https://www.nyc.gov/site/tlc/about/tlc-trip-record-data.page'
response = requests.get(url)

soup = BeautifulSoup(response.content, 'html.parser')

yellow_taxi_pattern = re.compile(r'.*yellow_tripdata_202[0-4]-(0[1-9]|1[0-2])\.parquet$', re.IGNORECASE)
hvfhv_pattern = re.compile(r'.*fhvhv_tripdata_202[0-4]-(0[1-9]|1[0-2])\.parquet$', re.IGNORECASE)

links = soup.find_all('a', href=True)

# Filter the links
yellow_taxi_links = [link['href'] for link in links if yellow_taxi_pattern.match(link['href'])]
hvfhv_links = [link['href'] for link in links if hvfhv_pattern.match(link['href'])]

# Use directory to store the downloaded Parquet files
os.makedirs('nyc_taxi_data', exist_ok=True)

# Function to download the Parquet files
def download_parquet_files(links, ride_type):
    for link in links:
        filename = link.split('/')[-1]
        filepath = os.path.join('nyc_taxi_data', filename)
        print(f"Downloading {ride_type} data: {filename}")
        parquet_response = requests.get(link)
        with open(filepath, 'wb') as f:
            f.write(parquet_response.content)
        print(f"Saved to {filepath}")

# Download Yellow Taxi and HVFHV Parquet files
download_parquet_files(yellow_taxi_links, 'Yellow Taxi')
download_parquet_files(hvfhv_links, 'HVFHV')
```
